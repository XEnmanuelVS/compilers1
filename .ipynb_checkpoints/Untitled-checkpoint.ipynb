```python
from cmp.automata import State
```

```python
states = [
    State(0, final=True),
    State(1),
    State(2),
]

states[0].add_transition('a', states[1])
states[1].add_transition('a', states[2])
states[2].add_transition('b', states[1])
states[1].add_transition('b', states[0])

start = states[0]

start
```

```python
def build_adj_list(cur, adj_list, visited=set()):
    amount = 1
    visited.add(cur.state)
    for symbol, nstates in cur.transitions.items():
        for nstate in nstates:
            if (cur.state == 1 and nstate.state == 2):
                print('*')
            if not adj_list.get(cur.state):
                adj_list[cur.state] = {}
            
            if not adj_list[cur.state].get(nstate.state):
                adj_list[cur.state] = {nstate.state: symbol}
            else:
                adj_list[cur.state][nstate.state] += "|" + symbol
            
            if not nstate.state in visited:
                visited.add(nstate.state)
                amount += build_adj_list(nstate, adj_list, visited)
    
    return amount
```

```python
adj_list = {}
N = build_adj_list(start, adj_list)
```
