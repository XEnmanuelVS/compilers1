```python
def append_or(s, new_s):
    # append or verifying that does not exists in s
    append = True
    for sub in s.split('|'):
        if sub == new_s:
            append = False
    if append:
        return s + "|" + new_s
    return s
```

```python
def build_adj_list(cur, adj_list, radj_list, visited=set()):
    amount = 1
    visited.add(cur.state)
    for symbol, nstates in cur.transitions.items():
        for nstate in nstates:
            if not adj_list.get(cur.state):
                adj_list[cur.state] = {}
            
            if not r_adj_list.get(nstate.state):
                r_adj_list[nstate.state] = {}
            
            if not adj_list[cur.state].get(nstate.state):
                adj_list[cur.state][nstate.state] = symbol
            else:
                adj_list[cur.state][nstate.state] = append_or(adj_list[cur.state][nstate.state], symbol)
            
            if not r_adj_list[nstate.state].get(cur.state):
                r_adj_list[nstate.state][cur.state] = symbol
            else:
                adj_list[nstate.state][cur.state] = append_or(adj_list[nstate.state][cur.state], symbol)
            
            if not nstate.state in visited:
                visited.add(nstate.state)
                amount += build_adj_list(nstate, adj_list, visited)
    
    return amount
```
